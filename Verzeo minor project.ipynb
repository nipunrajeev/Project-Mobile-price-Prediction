# Spliting into Train and Test set

```python
X=data.drop('price_range',axis=1)
X
```

```python
Y=data['price_range']
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
X_train
```

```python
Y_train
```

### Now we have split the data into train and test set where train set contains 1400 data and test 600 data.We first run models on x_train and y_train then make the model to predict price for x_test and then compare it with y_test to find the accuracy.... 

# 1) Logistic Regression

```python
#### using Standard Scaler ...
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()

X_train_std=std.fit_transform(X_train)

X_test_std=std.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train_std,Y_train)
```

```python
Y_pred=lr.predict(X_test_std)

Y_pred
```

```python
from sklearn.metrics import accuracy_score
lr_ac=accuracy_score(Y_test,Y_pred)

lr_ac
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_pred,Y_test)
```

# 2) KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train_std,Y_train)
```

```python
Y_pred=knn.predict(X_test_std)
Y_pred
```

```python
knn_ac=accuracy_score(Y_test,Y_pred)
knn_ac
```

```python
confusion_matrix(Y_pred,Y_test)
```

# SVM Classifier with linear and rbf kernel

```python
from sklearn.svm import SVC
linear=SVC(kernel="linear")
linear.fit(X_train,Y_train)
```

```python
Y_pred = linear.predict(X_test)
Y_pred
```

```python
linear_ac=accuracy_score(Y_test, Y_pred)
linear_ac
```

```python
confusion_matrix(Y_pred,Y_test)
```

```python
rbf=SVC(kernel="rbf")
rbf.fit(X_train,Y_train)
Y_pred = rbf.predict(X_test)
Y_pred
```

```python
rbf_ac=accuracy_score(Y_test, Y_pred)
rbf_ac
```

```python
confusion_matrix(Y_pred,Y_test)
```

# Classification report for Models.

```python
plt.bar(x=['knn','lr','linear','rbf'],height=[knn_ac,lr_ac,linear_ac,rbf_ac])
plt.xlabel("Algorithms")
plt.ylabel("Accuracy Score")
plt.show()
```

### From here we understand that best model for our data set is SVM Classifier with linear kernel
